```python
import copy
import numpy as np
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x, alpha):
    return np.maximum(alpha * x, x)

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.linspace(-6, 6, num=10000)

y = [None] * 3
y[0] = [sigmoid(v) for v in x]
y[1] = [tanh(v) for v in x]
y[2] = [relu(v, 0.1) for v in x]
```

```python
    
# plot 1
fig, ax1 = plt.subplots(1, 1, figsize=(5, 5))
ax1.spines['new_left'] = copy.copy(ax1.spines['left'])
ax1.spines['new_bottom'] = copy.copy(ax1.spines['bottom'])
ax1.spines['left'].set_position('center')
ax1.spines['bottom'].set_position(('data', 0))
ax1.xaxis.set_major_locator(MultipleLocator(5))
ax1.xaxis.set_minor_locator(MultipleLocator(1))
ax1.yaxis.set_major_locator(MultipleLocator(1))
ax1.yaxis.set_minor_locator(MultipleLocator(0.1))
ax1.set_xlim([-5.75, 5.75])
ax1.set_ylim([-0.1, 1.1])
ax1.set_xticks([-5, 5])
ax1.set_yticks([0, 0.5, 1])
ax1.set_yticklabels(['0', '0.5', '1'])
ax1.grid(which='major', linewidth=0.6, color='black', linestyle='-')
ax1.grid(which='minor', linewidth=0.3, color='grey', linestyle='-')
ax1.xaxis.set_tick_params(labelsize=12, zorder=10000)
ax1.yaxis.set_tick_params(labelsize=12, zorder=10000)
ax1.plot(x, y[0], linewidth=2)
plt.savefig('main/ch2-background/sigmoid.pdf', dpi=300, format='pdf')


# plot 2
fig, ax2 = plt.subplots(1, 1, figsize=(5, 5))
ax2.spines['new_left'] = copy.copy(ax2.spines['left'])
ax2.spines['new_bottom'] = copy.copy(ax2.spines['bottom'])
ax2.spines['left'].set_position('center')
ax2.spines['bottom'].set_position(('data', 0))
ax2.xaxis.set_major_locator(MultipleLocator(2))
ax2.xaxis.set_minor_locator(MultipleLocator(0.4))
ax2.yaxis.set_major_locator(MultipleLocator(2))
ax2.yaxis.set_minor_locator(MultipleLocator(0.4))
ax2.set_xlim([-2.3, 2.3])
ax2.set_ylim([-2.3, 2.3])
ax2.set_xticks([-2, 0, 2])
ax2.set_yticks([-2, 2])
ax2.grid(which='major', linewidth=0.6, color='black', linestyle='-')
ax2.grid(which='minor', linewidth=0.3, color='grey', linestyle='-')
ax2.xaxis.set_tick_params(labelsize=12, zorder=10000)
ax2.yaxis.set_tick_params(labelsize=12, zorder=10000)
ax2.plot(x, y[1], linewidth=2)
plt.savefig('main/ch2-background/tanh.pdf', dpi=300, format='pdf')


# plot 3
fig, ax3 = plt.subplots(1, 1, figsize=(5, 5))
ax3.spines['new_left'] = copy.copy(ax3.spines['left'])
ax3.spines['new_bottom'] = copy.copy(ax3.spines['bottom'])
ax3.spines['left'].set_position('center')
ax3.spines['bottom'].set_position(('data', 0))
ax3.xaxis.set_major_locator(MultipleLocator(1))
ax3.xaxis.set_minor_locator(MultipleLocator(0.2))
ax3.yaxis.set_major_locator(MultipleLocator(1))
ax3.yaxis.set_minor_locator(MultipleLocator(0.2))
ax3.set_xlim([-1.3, 1.3])
ax3.set_ylim([-1.3, 1.3])
ax3.set_xticks([-1, 0, 1])
ax3.set_yticks([-1, 1])
ax3.grid(which='major', linewidth=0.6, color='black', linestyle='-')
ax3.grid(which='minor', linewidth=0.3, color='grey', linestyle='-')
ax3.xaxis.set_tick_params(labelsize=12, zorder=10000)
ax3.yaxis.set_tick_params(labelsize=12, zorder=10000)
ax3.plot(x, y[2], linewidth=2)
plt.savefig('main/ch2-background/relu.pdf', dpi=300, format='pdf')
```
